```r
library(ggplot2)
library(ggsci)
library(RColorBrewer)
library(knitr)
```

```r
F1=read.csv("deepblink_f1score_final.csv")
F1_score=t(F1)
rownames(F1_score)=paste(rownames(F1_score),"deepblink",sep="")
#colnames(F1_score)=c("deepblink_F1")
F1_score1=data.frame(F1_score[grep(pattern = "^RCA_",rownames(F1_score)),])
F1_score2=data.frame(F1_score[grep(pattern = "^deepspot_",rownames(F1_score)),])
F1_score3=data.frame(F1_score[grep(pattern = "^deepblink_",rownames(F1_score)),])
F1_score4=data.frame(F1_score[grep(pattern = "^simfish_",rownames(F1_score)),])
colnames(F1_score1)=c("F1")
colnames(F1_score2)=c("F1")
colnames(F1_score3)=c("F1")
colnames(F1_score4)=c("F1")
F1_score1["class"]="RCA"
F1_score2["class"]="deepspot"
F1_score3["class"]="deepblink"
F1_score4["class"]="simfish"
F1_score1["class2"]="deepblink_RCA"
F1_score2["class2"]="deepblink_deepspot"
F1_score3["class2"]="deepblink_deepblink"
F1_score4["class2"]="deepblink_simfish"
```

```r
total1<-rbind(F1_score1,F1_score2,F1_score3,F1_score4)
total1$F1=as.numeric(total1$F1)
total1$F1
total1["class1"]="deepblink"
```

```r
F2=read.csv("RSFISH_f1score_final.csv")
F2_score=t(F2)
rownames(F2_score)=paste(rownames(F2_score),"RSFISH",sep="")
rownames(F2_score)
#colnames(F1_score)=c("deepblink_F1")
F2_score1=data.frame(F2_score[grep(pattern = "^RCA_",rownames(F2_score)),])
F2_score2=data.frame(F2_score[grep(pattern = "^deepspot_",rownames(F2_score)),])
F2_score3=data.frame(F2_score[grep(pattern = "^deepblink_",rownames(F2_score)),])
F2_score4=data.frame(F2_score[grep(pattern = "^simfish_",rownames(F2_score)),])
colnames(F2_score1)=c("F1")
colnames(F2_score2)=c("F1")
colnames(F2_score3)=c("F1")
colnames(F2_score4)=c("F1")
F2_score1["class"]="RCA"
F2_score2["class"]="deepspot"
F2_score3["class"]="deepblink"
F2_score4["class"]="simfish"
F2_score1["class2"]="RSFISH_RCA"
F2_score2["class2"]="RSFISH_deepspot"
F2_score3["class2"]="RSFISH_deepblink"
F2_score4["class2"]="RSFISH_simfish"
total2<-rbind(F2_score1,F2_score2,F2_score3,F2_score4)
total2$F1=as.numeric(total2$F1)
total2$F1
total2["class1"]="RSFISH"
```

```r
F3=read.csv("unet_f1score_final.csv")
F3_score=t(F3)
rownames(F3_score)=paste(rownames(F3_score),"unet",sep="")
F3_score
#colnames(F1_score)=c("deepblink_F1")
F3_score1=data.frame(F3_score[grep(pattern = "^RCA_",rownames(F3_score)),])
F3_score2=data.frame(F3_score[grep(pattern = "^deepspot_",rownames(F3_score)),])
F3_score3=data.frame(F3_score[grep(pattern = "^deepblink_",rownames(F3_score)),])
F3_score4=data.frame(F3_score[grep(pattern = "^simfish_",rownames(F3_score)),])
colnames(F3_score1)=c("F1")
colnames(F3_score2)=c("F1")
colnames(F3_score3)=c("F1")
colnames(F3_score4)=c("F1")
F3_score1["class"]="RCA"
F3_score2["class"]="deepspot"
F3_score3["class"]="deepblink"
F3_score4["class"]="simfish"
F3_score1["class2"]="unet_RCA"
F3_score2["class2"]="unet_deepspot"
F3_score3["class2"]="unet_deepblink"
F3_score4["class2"]="unet_simfish"
total3<-rbind(F3_score1,F3_score2,F3_score3,F3_score4)
total3$F1=as.numeric(total3$F1)
total3["class1"]="unet"
total<-rbind(total1,total2,total3)
```

```r
display.brewer.pal(12, 'Set3')
mycolor<-brewer.pal(12, 'Set3')
mycolor1<-brewer.pal(12, 'Set3')
totalmean<-aggregate(total$F1,by=list(total$class2),FUN=mean)#获取均值
totalsd<-aggregate(total$F1,by=list(total$class2),FUN=sd)
aov<-data.frame(totalmean,sd=totalsd$x)
colnames(aov)=c("model","mean","sd")
aov["group"]=c(rep('deepblink',4),rep("RSFISH",4),rep('unet',4))
aov["group1"]=rep(c("deepblink","deepspot","RCA","simfish"),3)
p=ggplot(data=aov,aes(x=group, y=mean,fill=group1))+
geom_bar(stat="identity",position="dodge",alpha=0.8)+
geom_errorbar(aes(ymax=mean+sd,ymin=mean-sd),position=position_dodge(0.9),width=0.15,color="black")+
scale_fill_manual(values=mycolor)+scale_color_manual(values=mycolor)+
 theme_bw() +
  theme(panel.grid.major=element_line(colour=NA),
        panel.background = element_rect(fill = "transparent",colour = NA),
        plot.background = element_rect(fill = "transparent",colour = NA),
        panel.grid.minor = element_blank())
p
```
