```r
library(ggplot2)
library(ggsci)
F1=read.csv("./f1score_test.csv")
F1_score=t(F1)
colnames(F1_score)=c("csv","F1")
F1_score1=data.frame(F1_score[grep(pattern = "^RCA_",rownames(F1_score)),])
F1_score2=data.frame(F1_score[grep(pattern = "^deepspot_",rownames(F1_score)),])
F1_score3=data.frame(F1_score[grep(pattern = "^deepblink_",rownames(F1_score)),])
F1_score4=data.frame(F1_score[grep(pattern = "^simfish_",rownames(F1_score)),])
F1_score1["class"]="RCA"
F1_score2["class"]="deepspot"
F1_score3["class"]="deepblink"
F1_score4["class"]="simfish"
total<-rbind(F1_score1,F1_score2,F1_score3,F1_score4)
total$F1=as.numeric(total$F1)
total$F1
```

```r
#基础绘图
mycolor<-pal_npg("nrc", alpha = 0.5)(8)
mycolor1<-pal_npg("nrc", alpha = 0.8)(8)
totalmean<-aggregate(total$F1,by=list(total$class),FUN=mean)#获取均值
totalsd<-aggregate(total$F1,by=list(total$class),FUN=sd)
aov<-data.frame(totalmean,sd=totalsd$x)
colnames(aov)=c("model","mean","sd")
ggplot(data=aov,aes(x=model, y=mean,fill=model,color=model))+
geom_bar(stat="identity",position="dodge")+
geom_errorbar(aes(ymax=mean+sd,ymin=mean-sd),position=position_dodge(0.9),width=0.15)+
scale_fill_manual(values=mycolor)+scale_color_manual(values=mycolor1)
```
